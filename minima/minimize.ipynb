### Ackley function

$f(x,y) = -A \exp{\left( -B \sqrt{\cfrac{x^2 + y^2}{2}} \right)} - \exp{\left( \cfrac{\cos{Cx} + \cos{Cy}}{2} \right)}$

$\argmin f(x,y) = (0, 0)$

https://www.sfu.ca/~ssurjano/ackley.html

```python
ackley = targets.ackley()
ackley.plot(256)
```

### Griewank function

$f(x,y) = \cfrac{x^2 + y^2}{A} - \cos{x} * \cos{\cfrac{y}{\sqrt2}}$

$\argmin f(x,y) = (0, 0)$

https://www.sfu.ca/~ssurjano/griewank.html

```python
griewank = targets.griewank()
griewank.plot(256)
```

### Rosenbrock function

$f(x,y) = A (y - x^2)^2 + (x - 1)^2$

$\argmin f(x,y) = (1, 1)$

https://www.sfu.ca/~ssurjano/rosen.html

```python
rosenbrock = targets.rosenbrock()
rosenbrock.plot(256)
```

### Schwefel function

$f(x,y) = -x \sin{\sqrt{|x|}} - y \sin{\sqrt{|y|}}$

$\argmin f(x,y) = (420.9687, 420.9687)$

https://www.sfu.ca/~ssurjano/schwef.html

```python
schwefel = targets.schwefel()
schwefel.plot(256)
```
